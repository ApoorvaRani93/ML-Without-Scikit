## Implementing Linear Regression with Gradient Descent

```python
import numpy as np
import pandas as pd
data = pd.read_csv("random.csv",header=None)
```

```python
df=data.values
```

```python
w= []
data1=[]
w1=[]
eeta = 0.0001
errtotprev=0.00
threshold = 0.0001
iter=0
w = np.zeros((1, df.shape[1]))
while(True):
    iter=iter+1
    sse=0
    X = df[:,0:(df.shape[1]-1)] 
    z=np.ones((len(X),1))
    X=np.append(z,X,axis=1)
    ypred = np.dot(X,w.T)
    y= df[:,-1].reshape(len(X),1)
    error = y - ypred
    sqer = error**2
    sse = sqer.sum()
    er = abs(sse - errtotprev)
    if er>threshold:
        errtotprev = sse
    else:
        break
    gradient = np.multiply(X,error).sum(axis=0)
    w = w+(eeta * gradient)
    print("iter",iter)
    print("%.4f"%sse)
    w1 = w
    w1 = [ round(item,4) for sublist in w1 for item in sublist]
```
